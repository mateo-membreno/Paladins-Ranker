```python
import matplotlib.pyplot as plt
import sqlite3 as sq
import pandas as pd

conn = sq.connect('paladins.db')


df = pd.read_sql_query('select * from player_data', conn)

win_counts = df[df['win_loss'] == 'W'].groupby('champion').size()
total_counts = df.groupby('champion').size()

proportions = win_counts / total_counts
print(type(proportions))
result_df = proportions.reset_index(name='proportion_of_wins')
print(type(result_df))
result_df.columns = ['champion', 'proportion_of_wins']
result_df = result_df.sort_values(by='proportion_of_wins', ascending=False)

plt.figure(figsize=(10, 12))  # Increase height


plt.barh(result_df['champion'], result_df['proportion_of_wins'], color='skyblue')  # Bar chart


plt.xlabel('Win %')
plt.ylabel('Champion', labelpad=15)
plt.title('Wins % by Champion')

plt.legend()
plt.show()



```

```python
import pandas as pd
import scipy.stats as stats
data = {
    'claim_amount': [5000, 12000, 7000, 15000, 3000, 20000, 25000, 4000],
    'is_fraudulent': [0, 1, 0, 1, 0, 1, 0, 0]
}

df = pd.DataFrame(data)

# Create a new column to categorize claim amounts
df['claim_over_10000'] = df['claim_amount'] > 10000

# Create a contingency table
contingency_table = pd.crosstab(df['claim_over_10000'], df['is_fraudulent'])
print(contingency_table)

# Perform the Chi-square test
chi2, p, dof, ex = stats.chi2_contingency(contingency_table)

print(f'Chi-square statistic: {chi2}')
print(f'P-value: {p}')
print(dof, ex)
```

```python
champion = 'Drogoz'
query = """SELECT *
        FROM player_data
        WHERE (match_id, win_loss) IN (
            SELECT match_id, win_loss
            FROM player_data
            WHERE champion = ?
        );"""
df = pd.read_sql_query(query, conn, params=(champion,))

total = df.groupby('champion').size()
won = df[df['win_loss'] == 'W'].groupby('champion').size()

win_p = won / total

result_df = win_p.reset_index(name='win_percentage')
print(type(result_df))
result_df.columns = ['champion', 'win_percentage']
result_df = result_df.sort_values(by='win_percentage')

plt.figure(figsize=(10, 12))  # Increase height

colors = ['skyblue' if champ != champion else 'orange' for champ in result_df['champion']]

plt.barh(result_df['champion'], result_df['win_percentage'], color=colors)  # Bar chart


plt.xlabel('Wins % by Champion')
plt.ylabel('Champion', labelpad=15)
plt.title(f'Win % when playing with {champion} on team')

plt.legend()
plt.show()



# need to select match IDS where drogoz is in match
# BUT select team that doesn't match drogoz's team
```
